```python
import numpy as np

# cópia de df
df_competencia = df

# Gera dados faltante no dataset
for col in df_competencia.columns[:-1]:
    df_competencia.loc[np.random.choice(df_competencia.index, 5), col] = np.nan
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Gráfico de linha para a média do comprimento da sépala de cada espécie
# Contagem de observações em cada categoria
count = df_competencia['genero'].value_counts()

# Crie um gráfico de barras
count.plot(kind='bar', color='skyblue')

# Configure os rótulos e título
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.title('Contagem de Observações por Gênero')

# Exiba o gráfico
plt.show()
```

```python
# Contagem de observações em cada categoria
count = df_competencia['genero'].value_counts()

# Crie um gráfico de setores (gráfico de pizza)
count.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue', 'lightgreen'])

# Configure o título
plt.title('Distribuição de Observações por Gênero')

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Contagem de valores únicos em 'competencia'
count = df['competencia'].value_counts()

# Crie um gráfico de barras
count.plot(kind='bar', color='skyblue')

plt.title('Contagem de Competências')
plt.xlabel('Competência')
plt.ylabel('Contagem')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Suponha que 'df' seja o seu DataFrame

# Crie um gráfico de barras para a contagem de valores únicos em 'deficiencia'
count = df['deficiencia'].value_counts()
plt.figure(figsize=(8, 6))
count.plot(kind='bar', color='skyblue')

# Adicione a contagem como texto no gráfico
plt.text(0.75, 0.9, f'Contagem: {count.sum()}', transform=plt.gca().transAxes, fontsize=12, color='black')

plt.title('Contagem de Valores Únicos em deficiencia')
plt.xlabel('Competência')
plt.ylabel('Contagem')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Suponha que 'df_competencia' seja o seu DataFrame

# Calcule a contagem de valores na coluna 'deficiencia' para cada valor na coluna 'competencia'
count_relationship = df_competencia.groupby('competencia')['deficiencia'].count()

# Crie um gráfico de barras para mostrar a relação entre 'competencia' e 'deficiencia'
count_relationship.plot(kind='bar', color='skyblue', figsize=(10, 6))

plt.title('Relação entre Competência e Deficiência')
plt.xlabel('Competência')
plt.ylabel('Contagem de Deficiência')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Suponha que 'df_competencia' seja o seu DataFrame

# Calcule a contagem de competências em cada categoria de cargo
count_relationship = df_competencia.groupby('cargo')['competencia'].count().reset_index()

# Crie um gráfico de dispersão
plt.scatter(range(len(count_relationship)), count_relationship['competencia'], c='skyblue', marker='o')

plt.title('Número de Competências em Cada Categoria de Cargo (Gráfico de Dispersão)')
plt.xlabel('Cargo')
plt.ylabel('Número de Competências')

# Defina os rótulos do eixo X a cada 10 categorias
plt.xticks(range(0, len(count_relationship), 10), count_relationship['cargo'][::10], rotation=90)

plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Suponha que 'df_competencia' seja o DataFrame que contém seus dados
# Certifique-se de que ele possui as colunas 'cargo' e 'deficiencia'

# Crie um DataFrame agregado com a contagem de deficiências por cargo
df_aggregated = df_competencia.groupby(['cargo'])['deficiencia'].count().reset_index()

# Ordene o DataFrame por contagem de deficiências em ordem decrescente
df_aggregated = df_aggregated.sort_values(by='deficiencia', ascending=False)

# Selecione apenas os 20 primeiros cargos com as contagens mais altas
df_aggregated = df_aggregated.head(20)

# Crie um gráfico de barras com a contagem de deficiências por cargo
plt.figure(figsize=(10, 6))
sns.barplot(data=df_aggregated, x='cargo', y='deficiencia')
plt.xlabel('Cargo')
plt.ylabel('Contagem de Deficiências')
plt.title('Contagem de Deficiências por Cargo')
plt.xticks(rotation=45, ha='right')

plt.show()


```
